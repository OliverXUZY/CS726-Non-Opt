```python
n=100
M = np.diag(np.repeat(2,n),k=0) + np.diag(np.repeat(-1,n-1),k=1) + np.diag(np.repeat(-1,n-1),k=-1)
b = np.zeros(n)
b[0] = 1

def f(x,m):
    return 1/2*M.dot(x).dot(x) - b.dot(x) + 1/2*m*x.dot(x)
def GD(x,m):
    #return M.dot(x) - b + m*x
    return x.dot(M) - b + m*x

m=0.1
x_star = lambda m: np.linalg.pinv(M + m*np.diag(np.ones(100))).dot(b)
f_star = f(x_star(m),m)
f_star
def minimum(m):
    x = np.linalg.inv(M+m*np.diag(np.ones(n))).dot(b)
    return x, f(x, m)
```

```python
L0 = max(np.linalg.eig(M)[0])
m0 = min(np.linalg.eig(M)[0])
L0, m0
```

### Algorithm

**Conjugate gradient**:
\
According to exact line search, we have:
$$
h_k = \frac{<\nabla f(x_k),p_k>}{<(M + \frac{m}{2})p_k,p_k>}
$$
\
$$
\beta_{k}=\frac{\left\|\nabla f\left(x_{k+1}\right)\right\|^{2}}{\left\langle\nabla f\left(x_{k+1}\right)-\nabla f\left(x_{k}\right), p_{k}\right\rangle}
$$
\
$$
p_{k+1}=\nabla f\left(x_{k+1}\right)-\beta_{k} p_{k}
$$

**Heavy-Ball**:
$$
\mathbf{x}_{k+1}=\mathbf{x}_{k}-\alpha_{1} \nabla f\left(\mathbf{x}_{k}\right)+\alpha_{2}\left(\mathbf{x}_{k}-\mathbf{x}_{k-1}\right)
$$
\
$$
\alpha_{1}=\frac{4}{(\sqrt{L}+\sqrt{m})^{2}} \text { and } \alpha_{2}=\left(\frac{\sqrt{L}-\sqrt{m}}{\sqrt{L}+\sqrt{m}}\right)^{2}
$$

```python
#### Nesterov:
def AGD(N,tol,m):
    x0 = np.repeat(0,n)
    L = 4 + m
    y0 = x0 - 1/L*GD(x0,m)
    v0 = y0
    value = [f(x0,m)]
    x_old = x0
    y_old = y0
    v_old = v0
    x_star,f_star = minimum(m)
    for i in np.arange(1,N+1):
        a = np.arange(1,i+1/2,1/2)
        A = a.sum()
        
        x_new = (A-a[-1])/A*y_old + a[-1]/A*v_old
        y_new = x_new - 1/L*GD(x_new,m)
        v_new = v_old - a[-1]/L*GD(x_new,m)
        
        value.append(f(y_new,m))
        x_old = x_new; y_old = y_new; v_old = v_new
        #if (f(y_new) - f_star)**2 <= tol:
        #    break
    return y_new, value

#### Nesterov for strongly smooth directly:
def AGD_ss(N,tol,m):
    x0 = np.repeat(0,n)
    L = 4 + m
    m0 = 4
    y0 = x0 - 1/L*GD(x0,m)
    #v0 = - 1/L*GD(x0,m)
    v0 = y0
    value = [f(x0,m)]
    x_old = x0
    y_old = y0
    v_old = v0
    lam = (m/L)**(1/2)
    a=[1]
    for i in np.arange(1,N+1):
        ak = lam/(1-lam)**i
        a.append(ak)
        A = sum(a)
        a_p = a[-1]*(m0 + m*(A-a[-1]))/(m0 + m*A)
        x_new = (A-a[-1])/(A-a[-1]+a_p)*y_old + a_p/(A-a[-1]+a_p)*v_old
        y_new = x_new - 1/L*GD(x_new,m)
        v_new = (m0 + m*(A-a[-1]))/(m0 + m*A)*v_old + m*a[-1]/(m0 + m*A)*x_new - a[-1]/(m0+m*A)*GD(x_new,m)
        
        value.append(f(y_new,m))
        x_old = x_new; y_old = y_new; v_old = v_new
        #if (f(y_new,m) - f_star,m)**2 <= tol:
        #    break
    return y_new, value

#### CG:
def CG(N,tol,m):
    x0 = np.repeat(0,n)
    x_old = x0
    p_old = GD(x0,m)
    value = [f(x0,m)]
    for i in range(N):
        h = GD(x_old,m).dot(p_old)/ (M.dot(p_old) + m/2*p_old).dot(p_old)
        x_new = x_old - h*p_old
        beta = GD(x_new,m).dot(GD(x_new,m))/(GD(x_new,m) - GD(x_old,m)).dot(p_old)
        p_new = GD(x_new,m) - beta*p_old
        
        value.append(f(x_new,m))
        x_old = x_new
        p_old = p_new
        
        #if (f(x_new,m) - f_star,m)**2 <= tol:
        #    break
    return x_new, value

#### Heavy-Ball:
def HB(N,tol,m):
    x0 = np.repeat(0,n)
    L = 4 + m
    alpha1 = 4/(L**(1/2) + m**(1/2))**2
    alpha2 = (L**(1/2) - m**(1/2))**2/(L**(1/2) + m**(1/2))**2
    value = [f(x0,m)]
    x1 = x0 - alpha1*GD(x0,m)
    value.append(f(x1,m))
    x_old = x1
    x_med = x1 - alpha1*GD(x1,m) + alpha2*(x1 - x0)
    for i in range(N):
        x_new = x_med - alpha1*GD(x_med,m) + alpha2*(x_med - x_old)        
        value.append(f(x_new,m))
        x_old = x_med
        x_med = x_new
        
        #if (f(x_new,m) - f_star,m)**2 <= tol:
        #    break
    return x_new, value
```

```python
def dfmake(N,m):
    f_star = f(x_star(m),m)
    #_,f_star = minimum(m)
    _,value_AGD = AGD(N,.001,m)
    _,value_AGDss = AGD_ss(N,0.1,m)
    _,value_CG = CG(N,0.01,m)
    _,value_HB = HB(N,0.01,m)
    #print(value_AGD[:N])

    value_AGD = value_AGD[:N]-f_star
    value_AGDss = value_AGDss[:N]-f_star
    value_CG = value_CG[:N]-f_star
    value_HB = value_HB[:N]-f_star
    df = pd.DataFrame({"Nesterov’s":value_AGD, "Nesterov’s for smooth and strongly convex": value_AGDss, 
                       "conjugate gradients": value_CG, "heavy ball":value_HB})
    return df
```

As we can see, The Nesterov's method has lowest convergence rate. The conjugate gradient method converge fastest when function has lower convexity parameter **m**. The heavy ball method has the faster convergence rate than nesterov's method. The direct Nesterov's method on strongly function converge faster than ordinary Nesterov's method.

#### Modify Nesterov

In each iteration:
$$
\begin{array}{c}\mathbf{x}_{k}=\frac{A_{k-1}}{A_{k}} \mathbf{y}_{k-1}+\frac{a_{k}}{A_{k}} \mathbf{v}_{k-1} \\ \mathbf{v}_{k}=\mathbf{v}_{k-1}-a_{k} \nabla f\left(\mathbf{x}_{k}\right) / L \\ \mathbf{y}_{k}=\mathbf{x}_{k}-\frac{1}{L} \nabla f\left(\mathbf{x}_{k}\right)\end{array}
$$
Let
$$
A = \{x_k-\frac{1}{L} \nabla f\left(\mathbf{x}_{k}\right),y_{k-1}- \frac{1}{L} \nabla f\left(\mathbf{y}_{k-1}\right)\}
\\
y_k = argmin_{y \in A} f(y)
$$
\
Then we have the theoretical iterations to achieve:
$$
f\left(\mathbf{x}_{k}\right)-f\left(\mathbf{x}^{*}\right) \leq \epsilon
$$
where:
$$
K = min\{O(\sqrt{\frac{L(\|x_0-x^*\|^2)}{\epsilon}}),O(\frac{L}{m}log(\frac{L(\|x_0-x^*\|^2)}{\epsilon}))\}
$$

```python
#### Nesterov:
def AGD(N,tol,m):
    x0 = np.repeat(0,n)
    L = 4 + m
    y0 = x0 - 1/L*GD(x0,m)
    v0 = y0
    value = [f(x0,m)]
    x_old = x0
    y_old = y0
    v_old = v0
    x_star,f_star = minimum(m)
    for i in np.arange(1,N+1):
        a = np.arange(1,i+1/2,1/2)
        A = a.sum()
        
        x_new = (A-a[-1])/A*y_old + a[-1]/A*v_old
        if f(x_new - 1/L*GD(x_new,m),m) < f(y_old - 1/L*GD(y_old,m),m):
            y_new = x_new - 1/L*GD(x_new,m)
        else:
            y_new = y_old - 1/L*GD(y_old,m)
        v_new = v_old - a[-1]/L*GD(x_new,m)
        
        value.append(f(y_new,m))
        x_old = x_new; y_old = y_new; v_old = v_new
        #if (f(y_new) - f_star)**2 <= tol:
        #    break
    return y_new, value

#### Nesterov for strongly smooth directly:
def AGD_ss(N,tol,m):
    x0 = np.repeat(0,n)
    L = 4 + m
    m0 = 4
    y0 = x0 - 1/L*GD(x0,m)
    #v0 = - 1/L*GD(x0,m)
    v0 = y0
    value = [f(x0,m)]
    x_old = x0
    y_old = y0
    v_old = v0
    lam = (m/L)**(1/2)
    a=[1]
    for i in np.arange(1,N+1):
        ak = lam/(1-lam)**i
        a.append(ak)
        A = sum(a)
        a_p = a[-1]*(m0 + m*(A-a[-1]))/(m0 + m*A)
        x_new = (A-a[-1])/(A-a[-1]+a_p)*y_old + a_p/(A-a[-1]+a_p)*v_old
        y_new = x_new - 1/L*GD(x_new,m)
        v_new = (m0 + m*(A-a[-1]))/(m0 + m*A)*v_old + m*a[-1]/(m0 + m*A)*x_new - a[-1]/(m0+m*A)*GD(x_new,m)
        
        value.append(f(y_new,m))
        x_old = x_new; y_old = y_new; v_old = v_new
        #if (f(y_new,m) - f_star,m)**2 <= tol:
        #    break
    return y_new, value

#### CG:
def CG(N,tol,m):
    x0 = np.repeat(0,n)
    x_old = x0
    p_old = GD(x0,m)
    value = [f(x0,m)]
    for i in range(N):
        h = GD(x_old,m).dot(p_old)/ (M.dot(p_old) + m/2*p_old).dot(p_old)
        x_new = x_old - h*p_old
        beta = GD(x_new,m).dot(GD(x_new,m))/(GD(x_new,m) - GD(x_old,m)).dot(p_old)
        p_new = GD(x_new,m) - beta*p_old
        
        value.append(f(x_new,m))
        x_old = x_new
        p_old = p_new
        
        #if (f(x_new,m) - f_star,m)**2 <= tol:
        #    break
    return x_new, value

#### Heavy-Ball:
def HB(N,tol,m):
    x0 = np.repeat(0,n)
    L = 4 + m
    alpha1 = 4/(L**(1/2) + m**(1/2))**2
    alpha2 = (L**(1/2) - m**(1/2))**2/(L**(1/2) + m**(1/2))**2
    value = [np.float64(f(x0,m))]
    x1 = x0 - alpha1*GD(x0,m)
    value.append(f(x1,m))
    x_old = x1
    x_med = x1 - alpha1*GD(x1,m) + alpha2*(x1 - x0)
    for i in range(N):
        x_new = x_med - alpha1*GD(x_med,m) + alpha2*(x_med - x_old)        
        value.append(np.float64(f(x_new,m)))
        x_old = x_med
        x_med = x_new
        
        #if (f(x_new,m) - f_star,m)**2 <= tol:
        #    break
    return x_new, value
```

```python
def f2(x,m):
    if x<1:
        return 25*x**2/2
    elif 1<=x<2:
        return x**2/2+24*x-12
    else:
        return 25*x**2/2-24*x+36

def GD2(x,m):
    if x<1:
        return 25*x
    elif 1<=x<2:
        return x+24
    else:
        return 25*x-24
```

```python
#### Nesterov:
def AGD2(N,tol,m):
    x0 = 3.3
    L = 25
    y0 = x0 - 1/L*GD2(x0,m)
    v0 = y0
    value = [f2(x0,m)]
    x_old = x0
    y_old = y0
    v_old = v0
    for i in np.arange(1,N+1):
        a = np.arange(1,i+1/2,1/2)
        A = a.sum()
        
        x_new = (A-a[-1])/A*y_old + a[-1]/A*v_old
        y_new = x_new - 1/L*GD2(x_new,m)
        v_new = v_old - a[-1]/L*GD2(x_new,m)
        
        value.append(f2(y_new,m))
        x_old = x_new; y_old = y_new; v_old = v_new
    return y_new, value

#### Heavy-Ball:
def HB2(N,tol,m):
    x0 = 3.3
    L = 25
    alpha1 = 4/(L**(1/2) + m**(1/2))**2
    alpha2 = (L**(1/2) - m**(1/2))**2/(L**(1/2) + m**(1/2))**2
    value = [f2(x0,m)]
    x1 = x0 - alpha1*GD2(x0,m)
    value.append(f2(x1,m))
    x_old = x1
    x_med = x1 - alpha1*GD2(x1,m) + alpha2*(x1 - x0)
    for i in range(N):
        x_new = x_med - alpha1*GD2(x_med,m) + alpha2*(x_med - x_old)        
        value.append(f2(x_new,m))
        x_old = x_med
        x_med = x_new
    return x_new, value
```

```python
def dfmake2(N,m):
    f_star = np.float64(f2(0,m))
    _,value_AGD = AGD2(N,.001,m)
    _,value_HB = HB2(N,0.01,m)

    value_AGD = value_AGD[:N]-f_star
    value_HB = value_HB[:N]-f_star
    df = pd.DataFrame({"Nesterov’s":value_AGD, "heavy ball":value_HB})
    return df
```

```python
df = dfmake2(100,1)
ax = df.plot.line(linewidth = 4,logy = True)
#ax.set_yticks([10**(-15),10**(-5),0.2])
None
```

As we can see, the heavy ball method has more fluctuations. The Nesterov's method converge to global minimizer immediately.
